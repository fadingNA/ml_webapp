```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('survey_results_public.csv')
```

```python
df.head()
```

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedComp"]]
df = df.rename({"ConvertedComp": "Salary"}, axis=1)
df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
print(df.columns)
```

```python
df = df[df["Employment"] == "Employed full-time"]
df = df.drop('Employment', axis=1)
df.info()
```

```python
df['Country'].value_counts()
```

```python
def concise_category(category, cutoff):
    category_map = {} # use dictionary
    for i in range(len(category)):
        if category.values[i] >= cutoff:
            category_map[category.index[i]] = category.index[i]
        else:
            category_map[category.index[i]] = 'Other'
    return category_map
```

```python
country_map = concise_category(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary ($USD) v Country')
plt.title('Salary Prediction')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 250000]
df = df[df["Salary"] >+ 10000]
df = df[df['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,7))
df.boxplot('Salary', 'Country', ax=ax)
plt.title('Salary after Clean')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_exp(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_exp)
```

```python
df["EdLevel"].unique()
```

```python
def clean_educa(x):
    bachelor = "Bachelor’s degree"
    master = "Master’s degree"
    profes = "Professional degree"

    if pd.isnull(x):
        return x

    if 'Bachelor’s degree' in str(x):
        return bachelor
    if 'Master’s degree' in str(x):
        return master
    if 'Professional degree' in str(x):
        return profes

    return x

df["EdLevel"] = df["EdLevel"].apply(clean_educa)
```
